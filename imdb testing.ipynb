```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

spark = (
            SparkSession
                .builder
    
                .appName("AndreasProject2IMDB")
                .master("local[4]")
    
                .config("spark.dynamicAllocation.enabled", "false")
                .config("spark.sql.adaptive.enabled", "false")    
                .getOrCreate()
        )

sc = spark.sparkContext

spark
```

```python
dfNameBasics = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\name.basics.tsv', sep=r'\t', inferSchema =True, header=True))
```

```python
dfNameBasics.show()
dfNameBasics.count()
```

```python
dfActors = (dfNameBasics.where(dfNameBasics.primaryProfession.contains('actor') | dfNameBasics.primaryProfession.contains('actress')))
dfActors = dfActors.drop('knownForTitles', 'primaryProfession')
dfActors.show()
dfActors.count()
```

```python
dfDirectors = (dfNameBasics.where(dfNameBasics.primaryProfession.contains('director')))
dfDirectors = dfDirectors.drop('knownForTitles','primaryProfession')
dfDirectors.show()
dfDirectors.count()
```

```python
dfTitleAkas = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\title.akas.tsv', sep=r'\t', inferSchema =True, header=True))
dfTitleAkas.printSchema()
dfTitleAkas.show()
```

```python
dfTitleBasics = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\title.basics.tsv', sep=r'\t', inferSchema =True, header=True))
dfTitleBasics.printSchema()
dfTitleBasics.show()
```

```python
dfFilms = (dfTitleBasics.where(dfTitleBasics.titleType.contains('movie')))
dfFilms.show()
```

```python
dfFilmsMod = dfFilms.drop('titleType','originalTitle', 'isAdult', 'endYear')
dfFilmsMod.show()
dfFilmsMod.count()
```

```python
dfTitleCrew = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\title.crew.tsv', sep=r'\t', inferSchema =True, header=True))
dfTitleCrew.printSchema()
dfTitleCrew.show()
```

```python
dfTitlePrincipals = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\title.principals.tsv', sep=r'\t', inferSchema =True, header=True))
dfTitlePrincipals.printSchema()
dfTitlePrincipals.show()
```

```python
dfTitlesActors = dfTitlePrincipals.drop('job','characters')
dfTitlesActors.show()
```

```python
dfTitleRatings = (spark.read.csv(r'C:\Users\Andreas\Downloads\imdb\title.ratings.tsv', sep=r'\t', inferSchema =True, header=True))
dfTitleRatings.printSchema()
dfTitleRatings.show()
```

```python
dfTitlesRatingsActors = (dfTitlesActors.join(dfTitleRatings, dfTitlesActors.tconst == dfTitleRatings.tconst, "inner").drop(dfTitleRatings.tconst))
dfTitlesRatingsActors.show()
```

```python
dfTitlesRatingsActors.count()
```

```python
import pyspark
dfTitlesRatingsActors.persist(pyspark.StorageLevel.MEMORY_AND_DISK)
```

```python
dfRatingsActorNames = (dfTitlesRatingsActors.join(dfActors, dfTitlesRatingsActors.nconst == dfActors.nconst, "inner").drop(
    dfActors.nconst))
dfRatingsActorNames.show()
```

```python
dfRatingsActorNames.count()
```

```python
dfRatingsActorsFilmNames = (dfRatingsActorNames.join(dfFilmsMod, dfRatingsActorNames.tconst == dfFilmsMod.tconst, "inner").drop(dfFilmsMod.tconst))
dfRatingsActorsFilmNames.show()
```

```python
import pyspark
dfRatingsActorsFilmNames.persist(pyspark.StorageLevel.MEMORY_AND_DISK)
```

```python
dfRatingsActorsFilmNames.count()
```

```python
dfCombinedDraft = dfRatingsActorsFilmNames.drop('birthYear','deathYear')
dfCombinedDraft.show()
```

```python
dfCombinedDraft.persist(pyspark.StorageLevel.MEMORY_AND_DISK)
```

```python
dfCombinedDraft.createOrReplaceTempView("CombinedDraft")
```

```python
dfCombinedDraft.rdd.getNumPartitions()
```

```python
filmsRatingsActors = dfCombinedDraft.coalesce(1)
```

```python
filmsRatingsActors.write.option("header","true").mode("append").parquet(r'C:\Users\Andreas\Download\IMDB Data\filmsRatingsActors.parquet')
```
